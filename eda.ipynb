```python
def get_esg_scores(esg_scores, ticker):
    """
    return an ESG scores of the input ticker
    """
    stock_universe = df_universe
    try:
        return esg_scores.loc[:,('SCORE', slice('CSR Strategy Score', 'Workforce Score'), str(df_universe.loc[ticker, 'SYM_INST_ORGPERMID']))]
    except KeyError:
        print('KeyEror: Stock does not have ESG scores')
```

```python
get_esg_scores(df_esg_scores, '@ZENCO3')
```

- Code snippet to get ESG scores for a particular stock
- To change stock, refer to ORGPERMID number, which can be retrieved from stock universe dataset

```python
def get_fundamental_data(ticker, contains=''):
    data_columns = list(df_items[df_items.apply(lambda row: row.astype(str).str.contains(contains, case=False).any(), axis=1)].item)
    return df_funds.loc[:,['day', 'security']+data_columns]
```

```python
get_fundamental_data('@ZIGAI3', 'cash')
```
